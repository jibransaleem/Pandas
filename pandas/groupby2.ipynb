```python
import numpy as np
import pandas as pd
ipl = pd.read_csv('/content/deliveries.csv')
movie = pd.read_csv('/content/imdb-top-1000.csv')
movie
```

```python
#looping thorugh group
genre = movie.groupby('Genre')
for group,data in genre:
      print(data[(data['IMDB_Rating'].max())==(data['IMDB_Rating'])]['Series_Title'])
```

```python
#split (apply) combine stratigie
# we split using group by and then apply transformation and then combine all the res
#apply - > builtin fun
def func(group):
  return ((group['Series_Title'].str[0] == 'A').sum())
genre.apply(func)
```

```python
#rank movie a/c to thier groups imbd
def ranking(group):
   group['RANK'] = group['IMDB_Rating'].rank(ascending=False)
   return group
genre.apply(ranking)
```

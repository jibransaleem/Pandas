value_counts -> frequency  count of each ele in series , frequency count of each row

```python

data.value_counts()
```

```python
#find player which has won most  potm in final matches of ipl
finals = ipl[ipl['MatchNumber']=='Final']
finals['Player_of_Match'].value_counts().max()
#in special macthes not normal match
mask =~ipl['MatchNumber'].str.isdigit()
ipl[mask]['Player_of_Match'].value_counts().max()
#Percentgae of bat/ball After toss win
ipl.loc[:,'TossDecision'].value_counts().plot(kind ='pie')
# total matches played by indvidual team
teams  = ipl.loc[:,['Team1','Team2']]
teams['Team1'].value_counts() + teams['Team2'].value_counts()
```

sort_Values


APPLY : both series and Dataframe


```python
#sort movies acording to their imdb rating :  highest -> lowest
movie.sort_values('imdb_rating',ascending=False)
#if there is any Missing values and if sort it ,they go down , if yout want them on top use parameter : na_position = 'first'
emp =  pd.read_csv('/content/employee_salary_extended.csv')
emp.sort_values('Salary',ascending=False)#ir respect of ascending descending , they will be on bottom by default
emp.sort_values('Salary',na_position='first')
# sort movies to their dates , then acording to to lexical order
movie.sort_values(['year_of_release','title_x'])
#sort movies to their name by decending and acending to their release of their date
movie.sort_values('title_x',ascending=False).sort_values('year_of_release')
#or
movie.sort_values(['title_x','year_of_release'],ascending = [False,True])
```

rank()

Apply : to series only


```python
#by default ranking happens in ascending order
# higher score - > higher rank
#if want low rank value - > ascending =  False
runs['batmans_rank'] = runs['batsman_run'].rank(ascending=False)
runs.sort_values(['batmans_rank'])
# if i dont do ascending = False , so rank highest rank means hight value
# then when i sort it , in decending value the highest rank will come up both means same but difeerent perspective
```

**rename()**


The .rename() method in Pandas is used to rename labels of rows or columns in a DataFrame. You can specify the new labels using a dictionary where the keys are the current labels and the values are the new labels.

it does not make change in the orignal data frame

```python
data = {
    'iq' : [100,90,80,70],
    'marks' : [ 20,30,40,50],
    'lpa'  : [12,13,14,15]
}
st = pd.DataFrame(data)
st.rename(columns={'lpa':'Package'})
```

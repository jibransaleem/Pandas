```python
import pandas as pd
import numpy as np
movie = pd.read_csv('/content/imdb-top-1000.csv')
deliv  = pd.read_csv('/content/deliveries.csv')
```

```python
movie.head(1)
# group by works on categorical data
genre = movie.groupby('Genre')
genre # group by pandas object
```

```python
#using functions on genre
genre.sum()
```

```python
#top 3 genre with most earning
genre.sum().sort_values(['Gross'],ascending=False).head(3)
#genre with highest avg imd rating
genre['IMDB_Rating'].mean().sort_values(ascending=False).head(1)
#director with most popularity
movie.groupby('Director')['No_of_Votes'].sum().sort_values(ascending=False).head(1)
```

```python
#find highest rated movie of each genre:
genre.max()['Series_Title']
```

```python
#find number of movies done by each actor
movie.groupby('Star1')['Series_Title'].count()
```

```python
 #find total number of groups -> len
len(movie.groupby('Genre'))
```

```python
#total rows in each group
movie.groupby('Genre').size()

```

```python
#first movie of each groups form
genre.first()
#Last movie of each group
genre.last()
# nth movie of each group

genre.nth(6)
```

```python
#get films of particular genre
genre.get_group('Horror') # fast as it is already grouped

# or
movie[movie['Genre']=='Horror'] # takes time ,first filter than apply so slow
```
